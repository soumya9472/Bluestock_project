# Performance Analytics

## Objective
Analyze mutual fund performance using return metrics.

## Metrics Covered
- 1 Year Return
- Fund Ranking
- Top Performing Funds
- Performance Comparison

```python
import pandas as pd

perf = pd.read_csv("../data/processed/07_scheme_performance_clean.csv")

print("Shape:", perf.shape)
perf.head()
```

```python
print(perf.columns)
```

```python
top_funds = perf.sort_values(
    by="return_1yr_pct",
    ascending=False
).head(10)

top_funds
```

```python
import matplotlib.pyplot as plt

top_funds.plot(
    x="scheme_name",
    y="return_1yr_pct",
    kind="bar",
    figsize=(10,5)
)

plt.title("Top 10 Funds by 1 Year Return")
plt.ylabel("1 Year Return (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
